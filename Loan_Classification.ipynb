```python
from pycaret.classification import *
```

```python
import pandas as pd
```

```python
df=pd.read_csv('LoansTrainingSet.csv')
```

```python
mean_value1 = df['Credit Score'].mean()
df['Credit Score'].fillna(mean_value1, inplace=True)

mean_value2 = df['Years in current job'].mode()[0]
df['Years in current job'].fillna(mean_value2, inplace=True)

mean_value3 = df['Annual Income'].mean()
df['Annual Income'].fillna(mean_value3, inplace=True)

mean_value4 = df['Months since last delinquent'].mean()
df['Months since last delinquent'].fillna(mean_value4, inplace=True)

mean_value5 = df['Bankruptcies'].mean()
df['Bankruptcies'].fillna(mean_value5, inplace=True)

mean_value5 = df['Tax Liens'].mean()
df['Tax Liens'].fillna(mean_value5, inplace=True)
```

```python
setup(df, target="Loan Status", ignore_features=["Loan ID", "Customer ID"])
```

```python
best_model = compare_models()
```
